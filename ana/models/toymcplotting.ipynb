```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the configurations specified by your mentor
configurations = [
    "Nominal",
    "No R_max cut", 
    "No d_0 cut",
    "No pitch angle cut",
    "Loose pitch angle",
    "Loose R_max cut"
]

# Corrected data - updated the last three configurations to 99.9x range
# Format: [efficiency_value, error_value]
config_data = {
    "Nominal": [99.07, 0.012],
    "No R_max cut": [99.97, 0.010], 
    "No d_0 cut": [99.16, 0.032],
    "No pitch angle cut": [99.94, 0.012],  # Changed from 99.04 to 99.94
    "Loose pitch angle": [99.94, 0.012],   # Changed from 99.04 to 99.94
    "Loose R_max cut": [99.97, 0.011]      # Changed from 99.07 to 99.97
}

# Extract efficiencies and errors
efficiencies = [config_data[config][0] for config in configurations]
errors = [config_data[config][1] for config in configurations]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Create horizontal error bars
y_positions = np.arange(len(configurations))
bars = ax.errorbar(efficiencies, y_positions, xerr=errors, 
                  fmt='o', capsize=5, capthick=2, 
                  markersize=8, linewidth=2, 
                  color='steelblue', markerfacecolor='lightblue',
                  markeredgecolor='steelblue', markeredgewidth=2)

# Customize the plot
ax.set_yticks(y_positions)
ax.set_yticklabels(configurations)
ax.set_xlabel('Veto efficiency [%]', fontsize=12, fontweight='bold')
ax.set_title('Veto Efficiency - Configuration Comparison', fontsize=14, fontweight='bold')

# Set x-axis range to focus on the efficiency values
ax.set_xlim(98.8, 100.1)

# Add grid for better readability
ax.grid(True, alpha=0.3, linestyle='--', axis='x')

# Add efficiency values as text labels
for i, (eff, err) in enumerate(zip(efficiencies, errors)):
    ax.text(eff + 0.02, i, f'{eff:.2f}{err:.3f}%', 
            va='center', ha='left', fontsize=10, fontweight='bold')

# Invert y-axis to match the original plot style
ax.invert_yaxis()

# Add a subtle background color
ax.set_facecolor('#f8f9fa')

plt.tight_layout()
plt.savefig('veto_efficiency_comparison.png', dpi=300, bbox_inches='tight', 
            facecolor='white', edgecolor='none')
plt.show()

# Print the data table
print("Veto Efficiency Configuration Results:")
print("-" * 60)
print(f"{'Configuration':<25} {'Efficiency (%)':<15} {'Error (%)':<10}")
print("-" * 60)
for config in configurations:
    eff, err = config_data[config]
    print(f"{config:<25} {eff:<15.2f} {err:<10.3f}")

print("\nConfiguration Details:")
print("1. within_lhr_max OFF")
print("3. within_pitch_angle OFF") 
print("4. within_pitch_angle OFF; within_pitch_angle_loose ON")
print("5. within_lhr_max OFF; within_lhr_max_loose ON")
```
